```python
#ข้อที่ 2 จำนวนผู้เสียชีวิต(Death) ระหว่างวันที่ 1 - 2 เมษายน 2563
url = [url1, url2] #เลือก url1 และ url2
test = pd.concat(url) #เลือก url
test = pd.DataFrame(test, columns= ['Last_Update','Deaths']) #สร้าง Column ที่มีการอัพเดทล่าสุดและจำนวนผู้เสียชีวิต
test #ทำการเรียกดูผลลัพธ์
```

```python
#ข้อที่ 3 จำนวนผู้เสียชีวิต(Death) ประเทศไทย(Thailand) วันที่ 7 เมษายน 2563
url7 = pd.DataFrame(url7, columns= ['Country_Region','Last_Update','Deaths']) #ใช้ข้อมูลของ url7 และสร้าง Columns ที่มีประเทศ อัพเดทล่าสุด และจำนวนผู้เสียชีวิต
test = url7.loc[(url7['Country_Region'] == 'Thailand')] #ทำการเจาะจงว่า url7 ใน columns ของประเทศต้องเป็นประเทศไทย
test #ทำการเรียกดูผลลัพธ์
```

```python
#ข้อที่ 4 จำนวนผู้รักษาหาย(Recovered) ระหว่างวันที่ 1 - 2 เมษายน 2563
url = [url1, url2] #เลือกข้อมูลของ url1 และ url2
test = pd.concat(url)  #เลือกเป็น url
test = pd.DataFrame(test, columns= ['Last_Update','Recovered']) #ให้ columns แสดงอัพเดทล่าสุดและจำนวนผู้รักษาหายแล้ว
test #ทำการเรียกดูผลลัพธ์
```

```python
#ข้อที่ 5 จำนวนผู้ติดเชื้อ(Confirmed) รัฐ Wisconsin (Province_State) ในวันที่ 10 เมษายน 2563
url10 = pd.DataFrame(url10, columns= ['Province_State','Last_Update','Confirmed']) #ให้ url10 แสดง columns ที่มีรัฐ อัพเดทล่าสุด และผู้ติดเชื้อ
test = url10.loc[(url10['Province_State'] == 'Wisconsin')] #รัฐที่จะเลือกแสดงคือ Wisconsin
test #ทำการเรียกดูผลลัพธ์
```

```python
#ข้อที่ 6 จำนวนผู้ติดเชื้อ(Confirmed) จำนวนผู้รักษาหาย(Recovered) ประเทศไทย(Thailand) ในวันที่ 1 เมษายน 2563
url1 = pd.DataFrame(url1, columns= ['Country_Region','Last_Update','Confirmed','Recovered']) #ให้ url1 แสดง column ที่มีประเทศ อัพเดทล่าสุด ผู้ติดเชื้อ และผู้ที่รักษาหายแล้ว
test = url1.loc[(url1['Country_Region'] == 'Thailand')] #ให้เจาะจงประเทศที่จะแสดงก็คือประเทศไทย
test #ทำการเรียกดูผลลัพธ์
```

```python
#ข้อที่ 7  จำนวนผู้ติดเชื้อ(Confirmed) จำนวนผู้เสียชีวิต(Deaths) ระหว่างวันที่ 3 - 5 เมษายน 2563
url = [url3, url4, url5] #เลือก url3 url4 url5
test = pd.concat(url) #เลือก url
test = pd.DataFrame(test, columns= ['Last_Update','Confirmed','Deaths']) #สร้าง column ที่จะแสดงคือ อัพเดทล่าสุด จำนวนผู้ติดเชื้อ และจำนวนผู้เสียชีวิต
test #ทำการเรียกดูผลลัพธ์
```

```python
#ข้อที่ 8 จำนวนผู้ติดเชื้อ(Confirmed) ผู้รักษาหาย(Recovered) ผู้เสียชีวิต(Deaths) ประเทศ(Western Sahara) ในวันที่ 7 เมษายน 2563
url7 = pd.DataFrame(url7, columns= ['Country_Region','Last_Update','Confirmed','Recovered','Deaths']) #ให้ url7 สร้าง column ที่มีประเทศ อัพเดทล่าสุด จำนวนผู้ติดเชื้อ ผู้รักษาหายแล้ว และผู้เสียชีวิต
test = url7.loc[(url7['Country_Region'] == 'Western Sahara')] #เจาะจงประเทศ 'Western Sahara
test #ทำการเรียกดูผลลัพธ์
```
